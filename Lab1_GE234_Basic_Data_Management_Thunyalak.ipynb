
## **4. การแสดงผลข้อมูล**
### **4.1 สร้างกราฟด้วย Matplotlib**  5 คะแนน
สร้างกราฟแสดงประชากรของแต่ละประเทศในภูมิภาค ASEAN โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. ข้อมูลที่กำหนด
data = {
    'Country': ['Indonesia', 'Philippines', 'Vietnam', 'Thailand', 'Myanmar',
                'Malaysia', 'Cambodia', 'Laos', 'Singapore', 'Brunei'],
    'Population_K': [275719.9, 111572.3, 99461.7, 66090.0, 55770.2,
                     32698.1, 16843.3, 7442.8, 5637.0, 445.4]
}

df = pd.DataFrame(data)

# 2. แปลงหน่วยเป็นล้านคน
df['Pop_Millions'] = df['Population_K'] / 1000

# 3. สร้างกราฟแท่ง
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Country'], df['Pop_Millions'], color='skyblue', edgecolor='navy')

plt.title('ASEAN Population by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population (Millions)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 5. ใส่ตัวเลขด้านบน
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```


### **4.2 การรวมข้อมูลเชิงพื้นที่และตาราง**
รวมข้อมูล Shapefile กับข้อมูลประชากรเพื่อแสดงข้อมูลประชากรในแต่ละจังหวัด โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# 1. อ่านข้อมูล
geo_data = gpd.read_file('/content/drive/MyDrive/GE234_2 2569/province')
df = pd.read_csv('/content/drive/MyDrive/GE234_2 2569/stat_c2567.csv')

# 2. คลีนข้อมูล
geo_data['adm1_name1'] = geo_data['adm1_name1'].str.strip().str.replace('จังหวัด', '')
df['ชื่อ'] = df['ชื่อ'].str.strip().str.replace('จังหวัด', '')

# 3. รวมข้อมูล
merged_data = geo_data.merge(df, left_on='adm1_name1', right_on='ชื่อ')

# ทำความสะอาดและแปลงข้อมูลประชากรเป็นตัวเลข
# ตรวจสอบก่อนว่าคอลัมน์มีอยู่จริงและไม่ใช่ NaN
if 'จำนวนประชากรทั้งหมด' in merged_data.columns:
    # ลบเครื่องหมายจุลภาคออกก่อนแปลงเป็นตัวเลข
    merged_data['จำนวนประชากรทั้งหมด'] = merged_data['จำนวนประชากรทั้งหมด'].astype(str).str.replace(',', '', regex=False)
    # แปลงเป็น int (หรือ float หากมีทศนิยม)
    merged_data['จำนวนประชากรทั้งหมด'] = pd.to_numeric(merged_data['จำนวนประชากรทั้งหมด'], errors='coerce').fillna(0).astype(int)


# 4. แสดงผล
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='จำนวนประชากรทั้งหมด',
                 cmap='OrRd',
                 legend=True,
                 edgecolor='0.5', # ใส่เส้นขอบจังหวัดให้ชัดขึ้น
                 ax=ax)

# 5. เพิ่มตัวเลขประชากรลงในพื้นที่จังหวัด
for idx, row in merged_data.iterrows():
    # หาจุดกึ่งกลางของจังหวัดเพื่อวาง Text
    centroid = row['geometry'].centroid
    # พิมพ์ตัวเลข (ใส่คอมมาแยกหลักพันเพื่อความสวยงาม)
    ax.annotate(text=f"{row['จำนวนประชากรทั้งหมด']:,}",
                xy=(centroid.x, centroid.y),
                horizontalalignment='center',
                fontsize=8,
                color='black')

plt.title('Population by Province')
plt.axis('off') # ปิดแกน x, y เพื่อความสวยงาม
plt.show()
```

1. ลองเพิ่มประเทศใหม่ใน DataFrame พร้อมคำนวณความหนาแน่นประชากร (5 คะแนน)

```python
import pandas as pd

# 1. ข้อมูลเดิมจากที่คุณให้มา (เพิ่มคอลัมน์ Area_km2 เข้าไปด้วย)
data = {
    'Country': ['Indonesia', 'Philippines', 'Vietnam', 'Thailand', 'Myanmar',
                'Malaysia', 'Cambodia', 'Laos', 'Singapore', 'Brunei'],
    'Population_K': [275719.9, 111572.3, 99461.7, 66090.0, 55770.2,
                     32698.1, 16843.3, 7442.8, 5637.0, 445.4],
    'Area_km2': [1904569, 300000, 331210, 513120, 676578,
                 330803, 181035, 236800, 728, 5765]
}

df = pd.DataFrame(data)

# 2. เพิ่มข้อมูลอีก 3 ประเทศใหม่
new_countries = pd.DataFrame({
    'Country': ['Timor-Leste', 'Papua New Guinea', 'Australia'],
    'Population_K': [1340.0, 9119.0, 25750.0],
    'Area_km2': [14874, 462840, 7692024]
})

# รวมข้อมูลเข้าด้วยกัน
df = pd.concat([df, new_countries], ignore_index=True)

# 3. คำนวณความหนาแน่นประชากร (คนต่อตารางกิโลเมตร)
# สูตร: (ประชากรหน่วยพัน * 1000) / พื้นที่
df['Density_per_sqkm'] = (df['Population_K'] * 1000) / df['Area_km2']

# 4. แปลงหน่วยประชากรเป็นล้าน
df['Pop_Millions'] = df['Population_K'] / 1000

# แสดงผลเฉพาะคอลัมน์ที่สำคัญ
print(df[['Country', 'Pop_Millions', 'Area_km2', 'Density_per_sqkm']].to_string(index=False))
```

2. นำไฟล์ Shapefile ของพื้นที่อื่นมาแสดงผลใน GeoPandas (5 คะแนน)

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. อ่านไฟล์
province = gpd.read_file('/content/drive/MyDrive/GE234_2 2569/thailand234')
rivers = gpd.read_file('/content/drive/MyDrive/GE234_2 2569/river')

# --- ส่วนที่เพิ่มเพื่อแก้ไขปัญหา ---
# เช็คก่อนว่ามีข้อมูลไหม
print(f"Province rows: {len(province)}")
print(f"River rows: {len(rivers)}")

# แปลงระบบพิกัดให้ตรงกัน (Crucial Step!)
if province.crs != rivers.crs:
    rivers = rivers.to_crs(province.crs)
# ------------------------------

# 3. แสดงผล
fig, ax = plt.subplots(figsize=(10, 10))

# วาดชั้นแรก (จังหวัด) - ลองเปลี่ยนสีให้เข้มขึ้นเพื่อเช็คการมองเห็น
province.plot(ax=ax, color='lightgreen', edgecolor='black', alpha=0.5)

# วาดชั้นที่สอง (แม่น้ำ)
rivers.plot(ax=ax, color='blue', linewidth=1)

# ปรับให้แกนขยายตามข้อมูลที่มีทั้งหมด
ax.set_aspect('equal')
plt.title('Map with Multiple Layers')
plt.show()
```

3. ลองสร้างแผนที่ด้วยสีที่แตกต่างกันโดยใช้ค่าอื่น เช่น area หรือ density

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# 1. โหลดข้อมูล
province = gpd.read_file('/content/drive/MyDrive/GE234_2 2569/province')
df = pd.read_csv('/content/drive/MyDrive/GE234_2 2569/stat_c2567.csv')

# ตรวจสอบและตั้งค่า CRS ให้กับ province
if province.crs is None:
    print("GeoDataFrame 'province' has no CRS. Setting to EPSG:4326.")
    province = province.set_crs(epsg=4326, allow_override=True)

# 2. คลีนข้อมูลชื่อจังหวัดให้ตรงกัน
province['adm1_name1'] = province['adm1_name1'].str.strip().str.replace('จังหวัด', '')
df['ชื่อ'] = df['ชื่อ'].str.strip().str.replace('จังหวัด', '')

# 3. รวมข้อมูล
merged = province.merge(df, left_on='adm1_name1', right_on='ชื่อ')

# 4. คำนวณความหนาแน่น (Density)
# แปลงเป็น UTM (32647) เพื่อคำนวณพื้นที่เป็น ตร.กม.
merged['area_sqkm'] = merged.to_crs(epsg=32647).area / 10**6

# ตรวจสอบและแปลง 'จำนวนประชากรทั้งหมด' เป็นตัวเลข หากยังไม่ได้ทำ
if 'จำนวนประชากรทั้งหมด' in merged.columns:
    if merged['จำนวนประชากรทั้งหมด'].dtype == 'object':
        merged['จำนวนประชากรทั้งหมด'] = merged['จำนวนประชากรทั้งหมด'].str.replace(',', '', regex=False)
        merged['จำนวนประชากรทั้งหมด'] = pd.to_numeric(merged['จำนวนประชากรทั้งหมด'], errors='coerce').fillna(0).astype(int)

merged['density'] = merged['จำนวนประชากรทั้งหมด'] / merged['area_sqkm']

# 5. สั่งวาดแผนที่
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

# ใช้คอลัมน์ density ในการระบายสี
merged.plot(column='density',
            cmap='viridis_r', # สีเหลือง-ส้ม-แดง
            legend=True,
            edgecolor='black',
            linewidth=0.3,
            ax=ax)

plt.title('Population Density by Province 2567')
plt.axis('off')
plt.show()
```
